```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from xlwt import Workbook
wb=Workbook()
import pandas as pd
ws=wb.add_sheet('Data')
ws.write(0,3,'Address')
ws.write(0,4,'Website')
ws.write(0,5,'Contact')
row=1
driver=webdriver.Firefox()
df = pd.read_excel('securityinformed.xls')
driver = webdriver.Firefox()
for i in range(len(df['Links'])):
    try:
        driver.get(df['Links'][i])
        print(driver.current_url)
        try:
            address=driver.find_element(By.CLASS_NAME,'address-icon')
            #a=h.find_element(By.TAG_NAME,'a')
            #print(a.get_attribute('href'))

            print(address.text)
            add=address.text
            ws.write(row, 3, add)
        except:
            pass
        try:
            web=driver.find_element(By.CLASS_NAME,'web-icon')
            print(web.text)
            website=web.text
            ws.write(row, 4, website)
        except:
            pass
        try:
            contact=driver.find_element(By.CLASS_NAME,'contact-icon')
            print(contact.text)
            con=contact.text
            ws.write(row, 5, con)
        except:
            pass
        row=row+1
        print('___________________________________________________________________')
    except:
        pass
    wb.save('security_informed.xls')
    


```

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
wait = WebDriverWait(driver, 10)
linkedin_element = wait.until(EC.presence_of_element_located((By.CLASS_NAME, "a2a_button_linkedin")))

# Execute JavaScript to wait for dynamic changes, if needed
driver.execute_script("arguments[0].scrollIntoView();", linkedin_element)

# Extract the href attribute value
linkedin_link = linkedin_element.get_attribute("href")

print("LinkedIn Link:", linkedin_link)
```
