```python
#url='https://www.trec.texas.gov/apps/license-holder-search/?lic_name=&industry=Real+Estate&email=&city=&county=Bexar&zip=&display_status=active&lic_hp=&ws=687&license_search=Search'
from selenium import webdriver
from selenium.webdriver.common.by import By
from xlwt import Workbook
from bs4 import BeautifulSoup as b
from selenium.webdriver.chrome.options import Options
wb=Workbook()
ws=wb.add_sheet('links')
ws.write(0,0,'Links')
ws.write(0,1,'County')
row=1
# options = Options()
# options.add_argument('--headless')
# options.add_argument('window-size=1920x1080')

for i in range(1,398):#680
    try:
        print(i)
        print('__________________________________________________________________________________________________')
        driver=webdriver.Chrome()
        driver.maximize_window()
        driver.get(f'https://www.trec.texas.gov/apps/license-holder-search/?lic_name=&industry=Real+Estate&email=&city=&county=Denton+&zip=&display_status=active&lic_hp=&ws=643&license_search=Search&showpage={i}')
        page_source=driver.page_source
        soup=b(page_source,'html.parser')
        main=soup.find_all(class_='panel panel-info')
        for m in main:
            links=m.find_all('h5',class_='panel-title')
            for link in links:
                a=link.find('a')
                if a:
                    href=f"https://www.trec.texas.gov/apps/license-holder-search/{a.get('href')}"
                    print(href)
                    

            county_div = m.find_all(class_='field-fluid col-xs-6 col-sm-4 col-md-4 col-lg-2')
            for county in county_div:
                cn = county.find_all(class_='label-fluid')
                try:
                    for c in cn:
                        if 'County' in c.text:
                            vlu  = c.find_next_sibling()# Check if 'county' is in the text of the c object
                            value = vlu.text
                            if value=='BEXAR':
                                print(value)
                                ws.write(row,0,href)
                                ws.write(row,1,value)
                                
                            row=row+1
                except Exception as e:
                    print(f'inner{e}')
                    pass
    except Exception as ex:
        print(ex)
        pass
    wb.save('links_license_DENTON.xls')
    driver.quit()
```
