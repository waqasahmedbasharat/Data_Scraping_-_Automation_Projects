```python
import requests

headers = {
    'authority': 'stockist.co',
    'accept': '*/*',
    'accept-language': 'en-US,en;q=0.9',
    'origin': 'https://alivebody.com.au',
    'referer': 'https://alivebody.com.au/',
    'sec-ch-ua': '"Not.A/Brand";v="8", "Chromium";v="114", "Google Chrome";v="114"',
    'sec-ch-ua-mobile': '?0',
    'sec-ch-ua-platform': '"Windows"',
    'sec-fetch-dest': 'empty',
    'sec-fetch-mode': 'cors',
    'sec-fetch-site': 'cross-site',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36',
}

params = {
    'tag': 'u7658',
    'latitude': '-38.0632184653318',
    'longitude': '145.62929399999996',
    'filter_operator': 'and',
    'distance': '279.8800912911476',
}

response = requests.get('https://stockist.co/api/v1/u7658/locations/search', params=params, headers=headers)

datalist=[]
import json
dat = json.loads(response.text)
for data in dat['locations']:
    name=data['name']
    latitude=data['latitude']
    longitude=data['longitude']
    address=''
    address1=data['address_line_1']
    address2=data['address_line_2']
    fulladdress=data['full_address']
    if address1:
        address += address1

    if address2 is not None:
        if address:
            address += ", " + address2
        else:
            address += address2

    if fulladdress is not None:
        if address:
            address += ", " + fulladdress
        else:
            address += fulladdress
    city=data['city']
    state=data['state']
    postal=data['postal_code']
    country=data['country']
    phone=data['phone']
    website=data['website']
    #print(website)
    email=data['email']
    #print(email)
    desc=data['description']
    #print(desc)
    distance = str(data['distance']) + ' km'
    alldata = [name,latitude,longitude,address,city,state,postal,country,phone,desc,website,email,distance]
    datalist.append(alldata)
    
import pandas as pd 
columns = ['Name', 'Latitude', 'Longitude', 'Address', 'City', 'State', 'Postal Code', 'Country','Phone', 'Description','Website', 'Email','Distance']
df = pd.DataFrame(datalist, columns=columns)
filename = 'aliveBody5.csv'
df.to_csv(filename, index=False)
```
