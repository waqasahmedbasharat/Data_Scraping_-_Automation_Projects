```python
import re
import requests
from bs4 import BeautifulSoup
import openpyxl

# Open the Excel file
workbook = openpyxl.load_workbook('pet groomers.xlsx')
sheet = workbook.active

# Column containing the website URLs
website_column = 'G'

# Column to store the extracted emails
email_column = 'H'
# Iterate through the rows, starting from the second row
for row in range(2, sheet.max_row + 1):
    website = sheet[website_column + str(row)].value

    try:
        # Send a GET request to the website
        response = requests.get(website)
        response.raise_for_status()

        # Extract emails using BeautifulSoup and regex
        soup = BeautifulSoup(response.text, 'html.parser')
        emails = re.findall(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b', soup.get_text())

        # Save emails in the email column or empty string if not found
        sheet[email_column + str(row)].value = ', '.join(emails) if emails else ''

        print(f"Emails extracted for {website}: {', '.join(emails)}" if emails else "No emails found")
    except (requests.RequestException, ValueError) as e:
        print(f"Error accessing website {website}: {str(e)}")

# Save the modified Excel file
workbook.save('output.xlsx')
```
