```python
import pandas as pd
from seleniumwire import webdriver
from selenium.webdriver.common.by import By
from xlwt import Workbook
import time
wb=Workbook()
ws=wb.add_sheet('links')
ws.write(0,0,'links')
ws.write(0,1,'Name')
ws.write(0,2,'Address')
ws.write(0,3,'Phone')
row=1
from bs4 import BeautifulSoup
options = {
        'proxies' :{

        "http": 'http://umarkhalid31speedproxies:XQvtQAuJOTvt0NQU_country-UnitedStates@p.speedproxies.net:31112',
        "https": 'http://umarkhalid31speedproxies:XQvtQAuJOTvt0NQU_country-UnitedStates@p.speedproxies.net:31112',
        },}
df=pd.read_excel('doctors_link.xls')

for i in range(len(df['links'])):
    try:
        driver=webdriver.Firefox(seleniumwire_options=options)
        driver.get(f"https://www.doctolib.fr/{df['links'][i]}")
        time.sleep(1)
        source=driver.page_source
        soup=BeautifulSoup(source,'html.parser')
        link=f"https://www.doctolib.fr/{df['links'][i]}"
        print(link)
        ws.write(row,0,link)
        try:
            name=soup.find(class_='dl-profile-header-name-speciality')
            nm=name.find('h1')
            print(nm.text)
            nm=nm.text
            ws.write(row,1,nm)
        except:
            pass
        try:
            addresses=soup.find_all(class_='dl-profile-card-content')
            for address in addresses:
                if "Carte et informations d'accès" in address.text:
                    profile_text_div = address.find(class_='dl-profile-text')
                    # Extract the name
                    name = profile_text_div.find(class_='dl-profile-practice-name').text.strip()
                    # Extract the remaining address by iterating through the generator
                    remaining_address_parts = [part.strip() for part in profile_text_div.stripped_strings if part.strip() != name]
                    remaining_address = ' '.join(remaining_address_parts)
                    print(name)
                    print(remaining_address)
                    com_add=name+' '+remaining_address
                    ws.write(row,2,com_add)
        except:
            pass
        try:
            contacts=soup.find_all(class_='dl-profile-box')
            for contact in contacts:
                if "Coordonnées" in contact.text:
                    con=contact.find('div',class_='flex')
                    print(con.text)
                    con=con.text
                    ws.write(row,3,con)
        except:
            pass
    except Exception as ex:
        print(ex)
        pass
    row=row+1
    wb.save('doctors_details.xls')
    driver.quit()
```
