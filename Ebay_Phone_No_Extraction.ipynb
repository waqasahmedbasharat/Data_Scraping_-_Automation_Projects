```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from collections import defaultdict

driver = webdriver.Chrome()
# path = 'https://www.ebay/.co.uk/b/Refurbished/bn_7117653182'
driver.get("https://www.ebay.co.uk/e/special-events/roe-mobile-phones-smartwatches")

phone_dict = defaultdict(str)

while True:
    eachphone = driver.find_elements(By.CLASS_NAME, 's-item__link')
    phonename_elements = driver.find_elements(By.CLASS_NAME, 's-item__title')

    for i in range(len(eachphone)):
        phone_name = phonename_elements[i].text
        phone_link = eachphone[i].get_attribute('href')
        phone_dict[phone_name] = phone_link


    for phone_name, phone_link in phone_dict.items():
        print(f"Phone Name: {phone_name}")
        print(f"Phone Link: {phone_link}")
        print()

    nextbtn = driver.find_element(By.CLASS_NAME, 'pagination__next')
    if 'disabled' in nextbtn.get_attribute('class'):
        print("Breaked...................")
        break
    else:
        nextbtn.click()

    time.sleep(2)

driver.quit()
```
